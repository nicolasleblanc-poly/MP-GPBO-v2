```python
def ch2xy_nparray(rows, cols):
    """
    Create a ch2xy vector where elements are ordered by column first, then row.

    Args:
        rows (int): Number of rows.
        cols (int): Number of columns.

    Returns:
        numpy.ndarray: An array of coordinates [(row, col), ...].
    """
    # Generate all row and column indices
    coordinates = [(row, col) for col in range(cols) for row in range(rows)]
    return np.array(coordinates) # /10

def ch2xy_tensor(rows, cols):
    """
    Create a ch2xy vector where elements are ordered by column first, then row.

    Args:
        rows (int): Number of rows.
        cols (int): Number of columns.

    Returns:
        numpy.ndarray: An array of coordinates [(row, col), ...].
    """
    # Generate all row and column indices
    coordinates = [(row, col) for col in range(cols) for row in range(rows)]
    return torch.tensor(coordinates, dtype=torch.float32) #/10 #np.array(coordinates)

# Example usage
rows, cols = 10,10  # Define the number of rows and columns
ch2xy_np = ch2xy_nparray(rows, cols) 
print(ch2xy_np)
ch2xy_t = ch2xy_tensor(rows, cols) 
print(ch2xy_t)
```
