```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
from BO import BayesianOptimizer
from ObjectiveFunction import ObjectiveFunction
```

```python
def compute_noise(noise):
    ackley = ObjectiveFunction("Ackley", dim=2)
    search_space = ackley.create_input_space()
    response = ackley.generate_true_response(search_space)
    kappa = 3

    opt = BayesianOptimizer(search_space, response, noise_std=noise)

    print("Ackley")
    gpbo_ackley_0, baseline = opt.train(
        kappa,
        initial_points=1,
        repetitions=30,
        iterations=150,
        strategy="Vanilla BO",
    )
    baseline, _ = baseline

    mpbo_ackley_0, _ = opt.train(
        kappa,
        initial_points=20,
        repetitions=30,
        iterations=150,
        strategy="MP-BO",
        begin_strat=20,
        follow_baseline=baseline,
    )

    print("Michalewicz 2D")

    mich2D = ObjectiveFunction("Michalewicz", dim=2)
    search_space = mich2D.create_input_space()
    response = mich2D.generate_true_response(search_space)
    kappa = 4

    opt = BayesianOptimizer(search_space, response, noise_std=noise)

    gpbo_mich2D_0, baseline = opt.train(
        kappa,
        initial_points=1,
        repetitions=30,
        iterations=150,
        strategy="Vanilla BO",
    )
    baseline, _ = baseline

    mpbo_mich2D_0, _ = opt.train(
        kappa,
        initial_points=20,
        repetitions=30,
        iterations=150,
        strategy="MP-BO",
        begin_strat=20,
        follow_baseline=baseline,
    )

    print("Michalewicz 4D")

    mich4D = ObjectiveFunction("Michalewicz", dim=4)
    search_space = mich4D.create_input_space()
    response = mich4D.generate_true_response(search_space)
    kappa = 5

    opt = BayesianOptimizer(search_space, response, noise_std=noise)

    gpbo_mich4D_0, baseline = opt.train(
        kappa,
        initial_points=1,
        repetitions=30,
        iterations=150,
        strategy="Vanilla BO",
    )
    baseline, _ = baseline

    mpbo_mich4D_0, _ = opt.train(
        kappa,
        initial_points=20,
        repetitions=30,
        iterations=150,
        strategy="MP-BO",
        begin_strat=20,
        follow_baseline=baseline,
    )

    print("Hartmann 6D")
    hartmann = ObjectiveFunction("Hartmann", dim=6)
    search_space = hartmann.create_input_space()
    response = hartmann.generate_true_response(search_space)
    kappa = 5

    opt = BayesianOptimizer(search_space, response, noise_std=noise)

    gpbo_hart_0, baseline = opt.train(
        kappa,
        initial_points=1,
        repetitions=30,
        iterations=150,
        strategy="Vanilla BO",
    )
    baseline, _ = baseline

    mpbo_hart_0, _ = opt.train(
        kappa,
        initial_points=20,
        repetitions=30,
        iterations=150,
        strategy="MP-BO",
        begin_strat=20,
        follow_baseline=baseline,
    )

    return (
        gpbo_ackley_0,
        mpbo_ackley_0,
        gpbo_mich2D_0,
        mpbo_mich2D_0,
        gpbo_mich4D_0,
        mpbo_mich4D_0,
        gpbo_hart_0,
        mpbo_hart_0,
    )
```

```python
noise = 0
(
    gpbo_ackley_0,
    mpbo_ackley_0,
    gpbo_mich2D_0,
    mpbo_mich2D_0,
    gpbo_mich4D_0,
    mpbo_mich4D_0,
    gpbo_hart_0,
    mpbo_hart_0,
) = compute_noise(noise)
noise_0 = [
    gpbo_ackley_0,
    mpbo_ackley_0,
    gpbo_mich2D_0,
    mpbo_mich2D_0,
    gpbo_mich4D_0,
    mpbo_mich4D_0,
    gpbo_hart_0,
    mpbo_hart_0,
]

np.save("noise_0.npy", noise_0)
```

```python
plt.plot(gpbo_ackley_0.mean(0), label="GP-BO Ackley")
plt.plot(mpbo_ackley_0.mean(0), label="MP-BO Ackley")
plt.plot(gpbo_mich2D_0.mean(0), label="GP-BO Michalewicz 2D")
plt.plot(mpbo_mich2D_0.mean(0), label="MP-BO Michalewicz 2D")
plt.plot(gpbo_mich4D_0.mean(0), label="GP-BO Michalewicz 4D")
plt.plot(mpbo_mich4D_0.mean(0), label="MP-BO Michalewicz 4D")
plt.plot(gpbo_hart_0.mean(0), label="GP-BO Hartmann 6D")
plt.plot(mpbo_hart_0.mean(0), label="MP-BO Hartmann 6D")
plt.legend()
```

```python
noise_levels = ["$0\\%$", "$2.5\\%$", "$10\\%$", "$20\\%$", "$50\\%$"]
noise_levels = ["$0\\%$", "$2.5\\%$", "$20\\%$"]

font = {"weight": "normal", "size": 12}
matplotlib.rc("font", **{"family": "serif", "serif": ["Times New Roman"]})
matplotlib.rc("text", usetex=True)


def plot_regrets_noise(noises):
    fig, axes = plt.subplots(
        3,
        4,
        figsize=(8, 5),
        sharex=True,
        sharey=True,
        gridspec_kw={"wspace": 0.2, "hspace": 0.2},
    )
    fig.subplots_adjust(left=0.2, wspace=0.4, hspace=0.4)

    titles = ["Ackley", "Michalewicz 2D", "Michalewicz 4D", "Hartmann"]

    plt.xticks([0, 75, 150])
    plt.yticks([0, 0.5, 1])

    for i, noise in enumerate(noises):
        x = np.arange(1, noise.shape[2] + 1)

        axes[0, 0].annotate(
            "Noise\n level ($\\sigma_n$)",
            xy=(-0.5, 1.1),
            xycoords="axes fraction",
            ha="center",
            va="center",
            fontsize=12,
            rotation=0,
        )
        axes[i, 0].annotate(
            noise_levels[i],
            xy=(-0.6, 0.5),
            xycoords="axes fraction",
            ha="center",
            va="center",
            fontsize=12,
        )

        for j in range(noise.shape[0]):
            if j % 2 == 0:
                axes[i, j // 2].plot(
                    x, noise[j].mean(0), color="red", linestyle="--", label="Vanilla BO"
                )
                axes[i, j // 2].fill_between(
                    x,
                    noise[j].mean(0) + noise[j].std(0) / np.sqrt(noise.shape[1]),
                    noise[j].mean(0) - noise[j].std(0) / np.sqrt(noise.shape[1]),
                    alpha=0.2,
                    color="red",
                )
            else:
                axes[i, j // 2].plot(
                    x, noise[j].mean(0), color="tab:blue", label="MP-BO"
                )
                axes[i, j // 2].fill_between(
                    x,
                    noise[j].mean(0) + noise[j].std(0) / np.sqrt(noise.shape[1]),
                    noise[j].mean(0) - noise[j].std(0) / np.sqrt(noise.shape[1]),
                    alpha=0.2,
                    color="tab:blue",
                )

                axes[i, j // 2].set_ylim(-0.1, 1.1)
                axes[i, j // 2].set_yticklabels(["0.0", " ", "1.0"])
                axes[i, j // 2].set_xticklabels(["0", "", "150"])
                axes[i, j // 2].axvline(
                    x=20,
                    linestyle="dotted",
                    color="black",
                    label="$q^* = 20$",
                )
                axes[-1, j // 2].set_xlabel("\\textbf{Iteration}", labelpad=-5)
                axes[0, j // 2].set_title("\\textbf{" + titles[j // 2] + "}")

        axes[i, 0].set_ylabel("\\textbf{Regret}", labelpad=-5)

        handles, labels = axes[1, 1].get_legend_handles_labels()

        legend = fig.legend(
            handles,
            labels,
            loc="upper center",
            bbox_to_anchor=(0.5, 1.05),
            ncol=4,
            fontsize=14,
            frameon=False,
        )
        fig.savefig(
            "noise.pdf",
            bbox_inches="tight",
        )


noises = np.array(
    [
        noise_0,
        noise_0025,
        # noise_01,
        noise_02,
        # noise_05,
    ]
)
plot_regrets_noise(noises)
```
